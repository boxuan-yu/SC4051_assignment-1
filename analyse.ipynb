```python
import matplotlib.pyplot as plt
import numpy as np

# Even values of k (fat-tree requires k to be even)
k_values = np.arange(2, 100, 2)

# Server count formula for k-ary fat-tree
server_counts = (k_values ** 3) / 4

# Create plot
plt.figure()
plt.plot(k_values, server_counts, marker='o')

plt.xlabel("k (Ports per Switch)")
plt.ylabel("Number of Servers")
plt.title("Server Count vs k in k-ary Fat-Tree")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fat-tree requires k to be even
k_values = np.arange(2, 100, 2)

# Inter-pod equal-cost shortest paths
inter_pod_paths = (k_values / 2) ** 2

plt.figure()
plt.plot(k_values, inter_pod_paths, marker='o')

plt.xlabel("k (Ports per Switch)")
plt.ylabel("Inter-Pod Equal-Cost Shortest Paths")
plt.title("Inter-Pod Path Count vs k in k-ary Fat-Tree")

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# k must be even in a k-ary fat-tree
k_values = np.arange(2, 100, 2)

# Bisection bandwidth formula: k^3 / 8
bisection_bandwidth = (k_values ** 3) / 8

plt.figure()
plt.plot(k_values, bisection_bandwidth, marker='o')

plt.xlabel("k (Ports per Switch)")
plt.ylabel("Bisection Bandwidth (Units of Link Bandwidth)")
plt.title("Bisection Bandwidth vs k in k-ary Fat-Tree")

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Even k values (fat-tree requires k to be even)
k_values = np.arange(2, 100, 2)

# Each link is 10 Gbps
link_bandwidth = 10  # Gbps

# Bisection bandwidth formula: (k^3 / 8) * link_bandwidth
bisection_bandwidth_gbps = (k_values ** 3 / 8) * link_bandwidth

plt.figure()
plt.plot(k_values, bisection_bandwidth_gbps, marker='o')

plt.xlabel("k (Ports per Switch)")
plt.ylabel("Bisection Bandwidth (Gbps)")
plt.title("Bisection Bandwidth vs k (10 Gbps per Link)")

plt.grid(True)
plt.show()
```
